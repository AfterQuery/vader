```python
import pandas as pd
```

```python
path_example_cols = '../data/graded.xlsx'
df = pd.read_excel(path_example_cols, sheet_name=0, engine='openpyxl')
```

```python
df
```

```python
df_rem = df.loc[:, df.columns.str.endswith('-remediation')]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))  # Adjust size as needed
df_rem.hist()
plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
df_rem.mean()
```

```python
(df_rem.describe().loc[['mean', 'std', '25%', '50%', '75%']] / 5 * 100).round(2)
```
