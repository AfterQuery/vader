```python
import pandas as pd

# Define the mapping of CWE identifiers to their descriptive names
cwe_mapping = {
    "CWE-89": "SQL Injection",
    "CWE-79": "Cross-Site Scripting",
    "CWE-78": "OS Command Injection",
    "CWE-22": "Path Traversal",
    "CWE-601": "URL Redirection to Untrusted Site",
    "CWE-23": "Relative Path Traversal",
    "CWE-20": "Improper Input Validation",
    "CWE-918": "Server-Side Request Forgery",
    "CWE-352": "Cross-Site Request Forgery",
    "CWE-798": "Use of Hard-coded Credentials"
}

# Function to modify a single CWE string
def add_cwe_descriptions(cwe_string):
    if pd.isna(cwe_string) or cwe_string.strip() == "":
        return cwe_string  # Handle NaN or empty strings
    # Split the string into individual CWEs
    cwe_list = [cwe.strip() for cwe in cwe_string.split(",")]
    # Modify each CWE if it's in the mapping
    modified_cwe_list = []
    for cwe in cwe_list:
        if cwe in cwe_mapping:
            modified_cwe_list.append(f"{cwe} ({cwe_mapping[cwe]})")
        else:
            modified_cwe_list.append(cwe)
    # Rejoin the modified CWEs with commas
    return ",".join(modified_cwe_list)

# Apply the function to the CWE column
df["CWE"] = df["CWE"].apply(add_cwe_descriptions)

# Display the modified DataFrame
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn theme first
sns.set_theme(style="white")

# Set the font family to Times New Roman and default font size to 17 for all plot elements
# Try Times New Roman, with fallbacks to similar serif fonts if unavailable
plt.rcParams['font.family'] = ['Times New Roman', 'DejaVu Serif', 'serif']
plt.rcParams['font.size'] = 17

# ── explode into list column & keep only multi-language cases ────────────────
df["LangList"] = (
    df["language"]
      .str.split(", ")
      .apply(lambda lst: [s.strip().title() for s in lst if s.strip()])
)

multi_df = df[df["LangList"].apply(len) > 1]

# ── binary indicator matrix: rows = cases, cols = languages ─────────────────
binary = multi_df["LangList"].apply(
    lambda lst: pd.Series(1, index=lst)
).fillna(0).astype(int)

# ── co-occurrence matrix: count pairwise language overlaps ──────────────────
cooc = binary.T.dot(binary)

# ── sort languages alphabetically and re-draw heatmap ─────────────────────
langs_sorted = sorted(cooc.index)          # alphabetical list
cooc_sorted = cooc.loc[langs_sorted, langs_sorted]

plt.figure(figsize=(10, 10))
sns.heatmap(
    cooc_sorted,
    cmap="Blues",
    annot=True,
    fmt="d",  # integer format for counts
    annot_kws={"size": 17},  # Font size for annotations (counts in cells)
    linewidths=0.5,
    square=True,
    cbar_kws={"label": "Co-occurrence Count"}
)

plt.title("Language Co-occurrence Matrix", fontsize=17, pad=20)
plt.xticks(rotation=45, ha="right", fontsize=17)
plt.yticks(rotation=0, fontsize=17)
plt.xlabel("Language", fontsize=17)
plt.ylabel("Language", fontsize=17)

# Adjust the colorbar label font size
cbar = plt.gcf().axes[-1]  # Get the colorbar axis
cbar.set_ylabel("Co-occurrence Count", fontsize=17)

plt.tight_layout()
plt.show()
```

The heatmap "Language Co-occurrence Matrix" (Figure 7) visualizes the co-occurrence of vulnerabilities across pairs of languages in the VADER dataset. The matrix, ordered alphabetically across both axes for programming languages, displays co-occurrence counts in off diagonal values, with color intensity indicating counts. Javascript and Python co-occur often at 21 cases, followed by Python often pairing SQL at 8 cases. At 5 cases, HTML often co-occurs with Javascript and Javascript often co-occurs PhP. These co-occurences may be due to common tech stacks in frontend web development and backend database workflows.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ── add a “Number of Files per Case” flag: 1-file vs multi-file ────────────────────────
df["Number of Files per Case"] = df["num_files"].apply(lambda n: "Single file" if n == 1 else "Multi-file")

# ── explode languages so every row = one (case, language, category) tuple ──
lang_cat = (
    df.assign(Language=df["language"].str.split(","))   # list
      .explode("Language")                              # one row per language
      .assign(Language=lambda d: d["Language"].str.strip().str.title())
)

# ── count rows per (language, category) pair ───────────────────────────────
lang_counts_cat = (
    lang_cat.groupby(["Language", "Number of Files per Case"])
            .size()
            .reset_index(name="Count")
)

# keep languages in overall-frequency order for nicer x-axis
lang_order = (
    lang_counts_cat.groupby("Language")["Count"].sum()
                  .sort_values(ascending=False)
                  .index
)

# ── seaborn grouped bar plot (blue = single-file, orange = multi-file) ─────
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 4))
ax = sns.barplot(
    data=lang_counts_cat,
    x="Language",
    y="Count",
    hue="Number of Files per Case",
    order=lang_order,
    palette=["chocolate", "steelblue"],   # blue vs orange
    dodge=True
)

ax.set(
    title="Programming Language Frequency (Single vs Multi-file Cases)",
    xlabel="Language",
    ylabel="Number of Cases"
)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

import seaborn as sns
import matplotlib.pyplot as plt

# df is your DataFrame with columns: num_files, num_languages
sns.set_theme(style="whitegrid")

plt.figure(figsize=(6, 4))
ax = sns.scatterplot(
    data=df,
    x="num_files",
    y="num_languages",
    s=60,               # marker size
    alpha=1.0
)

ax.set(
    title="Files vs. Languages per Case",
    xlabel="Number of Files",
    ylabel="Number of Distinct Languages"
)

plt.tight_layout()
plt.show()

import seaborn as sns
import matplotlib.pyplot as plt

severity_order = ["Very Low", "Low", "Medium", "High", "Critical"]
palette = sns.color_palette("rocket_r", len(severity_order))  # light → dark

sns.set_theme(style="whitegrid")

plt.figure(figsize=(6, 4))
ax = sns.scatterplot(
    data=df,
    x="num_files",
    y="num_languages",
    hue="Severity",
    hue_order=severity_order,
    palette=palette,
    s=60,
    alpha=1.0
)

ax.set(
    title="Files vs. Languages per Case",
    xlabel="Number of Files",
    ylabel="Number of Distinct Languages"
)

plt.tight_layout()
plt.show()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# ── explode (language, severity) pairs ───────────────────────────────────
lang_sev = (
    df.assign(Language=df["language"].str.split(","))
      .explode("Language")
      .assign(Language=lambda d: d["Language"].str.strip().str.title())
)

# ── counts and Medium-based ratios ───────────────────────────────────────
counts = lang_sev.groupby(["Language", "Severity"]).size().unstack(fill_value=0)
counts["Medium"].replace(0, np.nan, inplace=True)          # avoid /0
ratios = (
    counts[["Very Low", "Low", "Medium", "High", "Critical"]]
    .div(counts["Medium"], axis=0)
    .reset_index()
    .melt(id_vars="Language", var_name="Severity", value_name="Ratio")
)

# ── palette: light → dark with severity ──────────────────────────────────
severity_order = ["Very Low", "Low", "Medium", "High", "Critical"]
palette = dict(zip(
    severity_order,
    sns.color_palette("rocket_r", len(severity_order))   # light to dark
))

# ── seaborn bar plot ─────────────────────────────────────────────────────
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(
    data=ratios,
    x="Language",
    y="Ratio",
    hue="Severity",
    hue_order=severity_order,
    palette=palette,
    ax=ax
)

ax.set(
    title="Severity-to-Medium Ratio per Language",
    xlabel="Language",
    ylabel="Ratio to Medium Cases"
)
plt.xticks(rotation=45, ha="right")

# legend outside right
ax.legend(title="Severity", bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0)

plt.tight_layout()
plt.show()

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ── bin num_files into 1, 2-5, 6-10, >10 ─────────────────────────────────
df["Number of Files"] = pd.cut(
    df["num_files"],
    bins=[0, 1, 5, 10, np.inf],
    labels=["1 file", "2–5 files", "6–10 files", ">10 files"],
    include_lowest=True,   # include num_files==1
    right=True             # upper edge inclusive (5, 10 included)
)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 4))
sns.countplot(
    data=df,
    x="Severity",
    hue="Number of Files",
    order=["Very Low", "Low", "Medium", "High", "Critical"]
)

plt.title("Severity Distribution by Number of Files")
plt.xlabel("Severity")
plt.ylabel("Number of Cases")
plt.tight_layout()
plt.show()
```
